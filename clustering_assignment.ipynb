```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('marketing_campaign.csv',  sep="\t")
data.head()
```

```python
data.info()
```

```python
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'],format = '%d-%m-%Y')
```

```python
data.describe()
```

```python
# Drop unusable columns
data.drop(["ID", "Z_Revenue", "Z_CostContact"], axis=1, inplace=True)
```

```python
data.isna().sum()
```

```python
# Assume median
med = data[["Income"]].median()
print(med)
data["Income"] = data["Income"].fillna(float(med))
```

```python
data.isna().sum()
```

```python
# Encode Education and Martial_Status
data["Marital_Status"].value_counts()
```

```python
data['Marital_Status'].replace(['Married', 'Together'], 1, inplace=True)
data['Marital_Status'].replace(['Single', 'Divorced', 'Widow', 'Alone', 'YOLO', 'Absurd'], 0, inplace=True)
```

```python
data["Education"].value_counts()
```

```python
data['Education'].replace(['Basic', '2n Cycle'], 0, inplace=True)
data['Education'].replace(['Graduation'], 1, inplace=True)
data['Education'].replace(['Master', 'PhD'], 2, inplace=True)
```

```python
# Replace the date-time data with the customer enrollment length (in days)
data['Latest_Day'] = data['Dt_Customer'].max()
data['Customer_days'] = (data['Latest_Day'] - data['Dt_Customer']).dt.days
#data['Customer_days'].head()
data.drop(["Dt_Customer", "Latest_Day"], axis=1, inplace=True)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data[[
    "Year_Birth", 
    "Income", 
    "Recency", 
    "Customer_days",
    "MntWines", 
    "MntFruits", 
    "MntMeatProducts", 
    "MntFishProducts", 
    "MntSweetProducts", 
    "MntGoldProds",
    "NumDealsPurchases",
    "NumWebPurchases",
    "NumCatalogPurchases",
    "NumStorePurchases",
    "NumWebVisitsMonth"]].corr(), annot=True, cmap="Blues", annot_kws={"fontsize":12}, fmt=".2f", vmin=-1)
plt.show()
```

```python
data.info()
```

```python
# Scale the data using robust scalar
from sklearn.preprocessing import RobustScaler

X = data.loc[:]
s = RobustScaler()
X_s = s.fit_transform(X)
X_s.shape
```

```python
# PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # Keep 2 principle components
pca.fit(X_s)
X_PCA = pca.transform(X_s)
X_PCA[:5]
```

```python
X_PCA.shape
```

```python
x = X_PCA[:, 0]
y = X_PCA[:, 1]

plt.figure(figsize=(10, 8))
plt.scatter(x, y)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# 3 clustering algorithms
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
# First benchmark k for Kmeans using elbow method
inertia = []
n_cluster = []
for k in range(1, 10):
    km = KMeans(n_clusters=k).fit(X_PCA)
    inertia.append(km.inertia_)
    n_cluster.append(k)
plt.figure(figsize=(6, 4))
plt.plot(n_cluster, inertia)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
# We will use k = 4 for Kmeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn.metrics import davies_bouldin_score

kmean_model = KMeans(n_clusters=4).fit(X_PCA)
labels_km = kmean_model.labels_
score1 = metrics.silhouette_score(X_PCA, labels_km, metric='euclidean')
score2 = davies_bouldin_score(X_PCA, labels_km)
print("Silhouette score, higher value indicates better defined clusters: ", score1)
print("Davies-Bouldin score, lower value indicates better cluster separation: ", score2)
```

```python
# Visualize Kmeans result
plt.figure(figsize=(10, 8))
c = ['red','green','blue','purple']
color= [c[l] for l in labels_km]
plt.scatter(x, y, color=color)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Then we use Agglomerative Clustering
agg1 = AgglomerativeClustering(n_clusters=4, linkage='single').fit(X_PCA)
labels_agg1 = agg1.labels_

print("Model name: single-linkage agglomerative clustering.")
score1 = metrics.silhouette_score(X_PCA, labels_agg1, metric='euclidean')
score2 = davies_bouldin_score(X_PCA, labels_agg1)
print("Silhouette score, higher value indicates better defined clusters: ", score1)
print("Davies-Bouldin score, lower value indicates better cluster separation: ", score2)
plt.figure(figsize=(10, 8))
c = ['red','green','blue','purple']
color= [c[l] for l in labels_agg1]
plt.scatter(x, y, color=color)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Try different linkage
agg2 = AgglomerativeClustering(n_clusters=4, linkage='average').fit(X_PCA)
labels_agg2 = agg2.labels_

print("Model name: average-linkage agglomerative clustering.")
score1 = metrics.silhouette_score(X_PCA, labels_agg2, metric='euclidean')
score2 = davies_bouldin_score(X_PCA, labels_agg2)
print("Silhouette score, higher value indicates better defined clusters: ", score1)
print("Davies-Bouldin score, lower value indicates better cluster separation: ", score2)
plt.figure(figsize=(10, 8))
c = ['red','green','blue','purple']
color= [c[l] for l in labels_agg2]
plt.scatter(x, y, color=color)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Try different linkage
agg3 = AgglomerativeClustering(n_clusters=4, linkage='ward').fit(X_PCA)
labels_agg3 = agg3.labels_

print("Model name: Ward-linkage agglomerative clustering.")
score1 = metrics.silhouette_score(X_PCA, labels_agg3, metric='euclidean')
score2 = davies_bouldin_score(X_PCA, labels_agg3)
print("Silhouette score, higher value indicates better defined clusters: ", score1)
print("Davies-Bouldin score, lower value indicates better cluster separation: ", score2)
plt.figure(figsize=(10, 8))
c = ['red','green','blue','purple']
color= [c[l] for l in labels_agg3]
plt.scatter(x, y, color=color)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Try different linkage
agg4 = AgglomerativeClustering(n_clusters=4, linkage='complete').fit(X_PCA)
labels_agg4 = agg4.labels_

print("Model name: complete-linkage agglomerative clustering.")
score1 = metrics.silhouette_score(X_PCA, labels_agg4, metric='euclidean')
score2 = davies_bouldin_score(X_PCA, labels_agg4)
print("Silhouette score, higher value indicates better defined clusters: ", score1)
print("Davies-Bouldin score, lower value indicates better cluster separation: ", score2)
plt.figure(figsize=(10, 8))
c = ['red','green','blue','purple']
color= [c[l] for l in labels_agg4]
plt.scatter(x, y, color=color)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# DBSCAN eps = 0.1
dbs = DBSCAN(eps=0.1, min_samples=2).fit(X_PCA)
labels_dbs = dbs.labels_

print(labels_dbs)

print("Model name: DBSCAN.")
score1 = metrics.silhouette_score(X_PCA, labels_dbs, metric='euclidean')
score2 = davies_bouldin_score(X_PCA, labels_dbs)
print("Silhouette score, higher value indicates better defined clusters: ", score1)
print("Davies-Bouldin score, lower value indicates better cluster separation: ", score2)
plt.figure(figsize=(10, 8))
c = ['red','green','blue','purple', 'pink', 'cyan']
color= [c[l] if l in [0, 1, 2, 3, 4, 5] else 'grey' for l in labels_dbs]
plt.scatter(x, y, color=color)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# DBSCAN eps = 0.5
dbs = DBSCAN(eps=0.5, min_samples=2).fit(X_PCA)
labels_dbs = dbs.labels_

print(labels_dbs)

print("Model name: DBSCAN.")
score1 = metrics.silhouette_score(X_PCA, labels_dbs, metric='euclidean')
score2 = davies_bouldin_score(X_PCA, labels_dbs)
print("Silhouette score, higher value indicates better defined clusters: ", score1)
print("Davies-Bouldin score, lower value indicates better cluster separation: ", score2)
plt.figure(figsize=(10, 8))
c = ['red','green','blue','purple', 'pink', 'cyan']
color= [c[l] if l in [0, 1, 2, 3, 4, 5] else 'grey' for l in labels_dbs]
plt.scatter(x, y, color=color)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
